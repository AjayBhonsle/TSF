## Prediction using Unsupervised ML
### From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually. 

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
# read file from Github repo
file = 'https://raw.githubusercontent.com/AjayBhonsle/TSF/master/Task%202/Iris.csv'
df1 = pd.read_csv(file)
df1
```

```python
# check for null values
df1.isnull().sum()
```

```python
# drop not necessary column
df2 = df1.drop(['Id'],axis='columns')
df2.columns
```

```python
x = df2[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
y = df2[['Species']].values
```

```python
# Finding the optimum number of clusters for k-means classification
sse = []
for i in range(1, 11):
    means = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    means.fit(x)
    sse.append(means.inertia_)

# locate the elbow point
kn = KneeLocator(range(1,11), sse, curve='convex', direction='decreasing')

# Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(1, 11), sse, 'bx-')
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.vlines(kn.elbow, plt.ylim()[0], plt.ylim()[1], linestyles='dashed')
plt.show()
```

```python
pair = sns.PairGrid(df2, hue="Species")
pair.map_offdiag(sns.scatterplot)
pair.add_legend()
```

### Model building

```python
km = KMeans(n_clusters = 3, init = 'k-means++',max_iter = 300, n_init = 10, random_state = 0)
k_pred = km.fit_predict(x)
```

```python
k_pred
```

```python
df3 = df2.copy()
df3['Cluster'] = k_pred
df3
```

```python
# Visualising the clusters - On the first two columns
plt.figure(figsize=(15,10))
plt.scatter(x[k_pred == 0, 0], x[k_pred == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[k_pred == 1, 0], x[k_pred == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[k_pred == 2, 0], x[k_pred == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:,1], s = 100, c = 'black', label = 'Centroids')
```
